"""Attribute information:

   Input variables (based on physicochemical tests):
   1 - fixed acidity (tartaric acid - g / dm^3)
   2 - volatile acidity (acetic acid - g / dm^3)
   3 - citric acid (g / dm^3)
   4 - residual sugar (g / dm^3)
   5 - chlorides (sodium chloride - g / dm^3
   6 - free sulfur dioxide (mg / dm^3)
   7 - total sulfur dioxide (mg / dm^3)
   8 - density (g / cm^3)
   9 - pH
   10 - sulphates (potassium sulphate - g / dm3)
   11 - alcohol (% by volume)
   Output variable (based on sensory data): 
   12 - quality (score between 0 and 10)

```python
#Removing the outliers from the dependent Attributes i.e. volatile.acidity, citric.acid, chlorides, total.sulfur.dioxide,
#sulphates, alcohol
#Removing outliers fixed.acidity
plt.figure(figsize=(10, 10))
boxplot = df.boxplot(column=['fixed.acidity'])
```

```python
#Whisker lenght = Q3+1.5*IQR
q75 = df["fixed.acidity"].quantile([.75])
q75 = q75.values[0]
q25 = df["fixed.acidity"].quantile([.25])
q25 = q25.values[0]
IQR = q75-q25
wlu = q75 + 1.5*IQR
wll = q25-1.5*IQR
print(wlu,wll)
```

```python
#so the values outside this will be considered as outliers
#removing these outliers will not help so replacing the outliers with median
#Calculating total number of outlier
df['fixed.acidity'][(df['fixed.acidity'] > wlu ) | (df['fixed.acidity'] < wll )].count()
```

```python
plt.figure(figsize=(10, 10))
sns.distplot(df['fixed.acidity'])
```

```python
print('Mean :', df['fixed.acidity'].mean())
print('Median :', df['fixed.acidity'].median())
print('Mode', df['fixed.acidity'].mode())
Acidity_Median = df['fixed.acidity'].mean()
```

```python
# Since the distribution is righ skewed and correct replacement will be Median
# So replacing the outliers with Median
df.loc[df['fixed.acidity']> wlu, 'fixed.acidity'] = np.nan # replacing the outliers with NAN values
df.loc[df['fixed.acidity']< wll, 'fixed.acidity'] = np.nan # replacing the outliers with NAN values
df.isna().sum()
#All the outliers
```

```python
df['fixed.acidity'].fillna(Acidity_Median,inplace=True)
df.isna().sum() # so now the values got replaced 
```

```python
# Plotting the Acidity Distribution and its better this time
plt.figure(figsize=(10, 10))
sns.distplot(df['fixed.acidity'])
```
